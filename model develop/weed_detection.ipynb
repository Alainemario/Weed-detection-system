```python
from ultralytics import YOLO
import cv2
import os

# Path to trained model
MODEL_PATH = "best.pt"

# Path to your input video
VIDEO_PATH = "..\dataset\weed detection simulation video - Made with Clipchamp.mp4"  # <--- Update this path

# Output folder
OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)
OUTPUT_PATH = os.path.join(OUTPUT_DIR, "output_video.mp4")

def main():
    # Load model
    model = YOLO(MODEL_PATH)

    # Open the video file
    cap = cv2.VideoCapture(VIDEO_PATH)

    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get video properties for the output writer
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object
    # 'mp4v' is a common codec for .mp4 files
    fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
    out = cv2.VideoWriter(OUTPUT_PATH, fourcc, fps, (width, height))

    print(f"Processing video... Press 'q' to stop early.")

    while cap.isOpened():
        success, frame = cap.read()

        if not success:
            break # End of video

        # Run inference on the frame
        # stream=True is recommended for videos to manage memory
        results = model(frame, conf=0.3, stream=True)

        # Iterate through results (usually just one per frame)
        for r in results:
            annotated_frame = r.plot() # Draws boxes + labels

            # Write the frame to the output video
            out.write(annotated_frame)

            # Display the frame in a window
            cv2.imshow("Weed Detection", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release everything when done
    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print(f"âœ… Processing complete. Output saved at: {OUTPUT_PATH}")

if __name__ == "__main__":
    main()
```

```python
from ultralytics import YOLO
import cv2
import os

# Path to trained model
MODEL_PATH = "best.pt"
VIDEO_PATH = "..\dataset\weed detection simulation video.mp4" # <--- Update this

# Output setup
OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)
OUTPUT_PATH = os.path.join(OUTPUT_DIR, "weed_count_output.mp4")

def main():
    # Load model
    model = YOLO(MODEL_PATH)

    cap = cv2.VideoCapture(VIDEO_PATH)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(OUTPUT_PATH, fourcc, fps, (width, height))

    # Set to store unique object IDs (avoids double counting)
    unique_weed_ids = set()

    print("Processing video... Press 'q' to stop.")

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Run Object Tracking (persist=True is required for tracking)
        results = model.track(frame, conf=0.3, persist=True)

        for r in results:
            # Check if there are any detections with IDs
            if r.boxes.id is not None:
                # Get the IDs as integers
                track_ids = r.boxes.id.int().cpu().tolist()
                
                # Add new IDs to our set
                for track_id in track_ids:
                    unique_weed_ids.add(track_id)

            # Draw the standard tracking visualization
            annotated_frame = r.plot()

            # (Optional) Overlay the count on the video
            cv2.putText(annotated_frame, f"Total Weeds: {len(unique_weed_ids)}", (20, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            out.write(annotated_frame)
            cv2.imshow("Weed Tracking", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print("------------------------------------------------")
    print(f"âœ… Processing complete.")
    print(f"ğŸŒ± Total Unique Weeds Detected: {len(unique_weed_ids)}")
    print(f"ğŸ“ Video saved at: {OUTPUT_PATH}")
    print("------------------------------------------------")

if __name__ == "__main__":
    main()
```

```python
from ultralytics import YOLO
import cv2
import os

# --- CONFIGURATION ---
MODEL_PATH = "best.pt"
VIDEO_INPUT = r"..\\weed_test_2.mp4"  # Update path if needed
OUTPUT_DIR = "video_output"
CROP_DIR = os.path.join(OUTPUT_DIR, "crops")

# Ensure directories exist
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(CROP_DIR, exist_ok=True)

def main():
    # 1. Load Model
    print(f"[INFO] Loading model: {MODEL_PATH}")
    model = YOLO(MODEL_PATH)

    # 2. Open Video Source
    cap = cv2.VideoCapture(VIDEO_INPUT)
    if not cap.isOpened():
        print(f"[ERROR] Could not open video: {VIDEO_INPUT}")
        return

    # Get video properties
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = int(cap.get(cv2.CAP_PROP_FPS))
    
    # 3. Setup Video Writer
    output_video_path = os.path.join(OUTPUT_DIR, "processed_weed_video.mp4")
    fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
    out_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    # Set to store unique object IDs to avoid double counting/cropping
    unique_weed_ids = set()

    print(f"[INFO] Processing video... Press 'q' to stop.")

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break
        
        # 4. Run Object Tracking 
        # persist=True is CRITICAL for tracking to work across frames
        results = model.track(frame, conf=0.3, persist=True, verbose=False)

        for r in results:
            # Visualize the tracking (draws boxes and IDs)
            annotated_frame = r.plot()
            
            # Check if any objects were detected and have IDs assigned
            if r.boxes.id is not None:
                # Get IDs and coordinates
                track_ids = r.boxes.id.int().cpu().tolist()
                boxes = r.boxes.xyxy.cpu().tolist()
                
                # Loop through detections in this frame
                for box, track_id in zip(boxes, track_ids):
                    
                    # 5. Check if this is a NEW weed we haven't seen before
                    if track_id not in unique_weed_ids:
                        unique_weed_ids.add(track_id)
                        
                        # --- CROP AND SAVE ONLY ONCE ---
                        x1, y1, x2, y2 = map(int, box)
                        
                        # Ensure coordinates are within frame bounds
                        y1, y2 = max(0, y1), min(height, y2)
                        x1, x2 = max(0, x1), min(width, x2)
                        
                        crop = frame[y1:y2, x1:x2]
                        
                        if crop.size > 0:
                            # Save with the unique ID in the filename
                            crop_filename = f"weed_id_{track_id}.jpg"
                            cv2.imwrite(os.path.join(CROP_DIR, crop_filename), crop)
                            print(f"[NEW] Weed ID {track_id} detected and cropped.")

            # 6. Display Unique Count on Screen
            cv2.putText(annotated_frame, f"Unique Weeds: {len(unique_weed_ids)}", (20, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            # 7. Write to Video File and Display
            out_writer.write(annotated_frame)
            cv2.imshow("Weed Tracking", annotated_frame)

        # Quit on 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Cleanup
    cap.release()
    out_writer.release()
    cv2.destroyAllWindows()
    
    print("------------------------------------------------")
    print(f"âœ… Processing complete.")
    print(f"ğŸŒ± Total Unique Weeds Detected: {len(unique_weed_ids)}")
    print(f"ğŸ“ Video saved at: {os.path.abspath(output_video_path)}")
    print(f"ğŸ“ Unique crops saved in: {os.path.abspath(CROP_DIR)}")
    print("------------------------------------------------")

if __name__ == "__main__":
    main()
```
