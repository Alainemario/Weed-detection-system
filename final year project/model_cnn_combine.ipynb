```python
from ultralytics import YOLO

# 1. Build the model from your custom YAML
# This initializes the weights from ResNet (backbone) and random weights for YOLO (head)
model = YOLO("yolov8-resnet50.yaml") 

# 2. Train on your agricultural dataset
# 'data.yaml' is your standard dataset config file
results = model.train(
    data="..\\dataset\\data.yamll",
    epochs=100,
    imgsz=640,
    batch=16,
    name="resnet_weed_detection"
)

# 3. Export for deployment (e.g., to ONNX for a tractor/drone)
model.export(format="onnx")
```
